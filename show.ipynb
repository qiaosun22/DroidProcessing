```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

path = "/media/sun/12TB/datasets/droid/droid_raw/processed_0421/AUTOLab/success/2023-11-28/Tue_Nov_28_08:46:19_2023/frames/ext1_depth_0.npz"
depth = np.load(path)['arr_0']
# depth.min(), depth.max()    
depth = np.nan_to_num(depth, nan=0)
np.clip(depth, 1e-2, 2e3, out=depth)
depth.min(), depth.max()    

# 保存 depth 图像为单通道
# Normalize depth to [0, 255] and convert to uint16
depth_normalized = (depth - depth.min()) / (depth.max() - depth.min()) * 255
depth_normalized = depth_normalized.astype(np.uint8)
cv2.imwrite("saved/depth.png", depth_normalized)

# 手动归一化到 [0, 255]
depth_min = depth.min()
depth_max = depth.max()

print(depth_min, depth_max)
depth_normalized = (depth - depth_min) / (depth_max - depth_min) * 255

# 转换为 uint8 类型
depth_normalized = depth_normalized.astype(np.uint8)

# 使用 Matplotlib 的 colormap 将深度图转换为三通道彩色图
depth_colored = plt.cm.viridis(depth_normalized)[:, :, :3]  # 使用 viridis colormap
depth_colored = (depth_colored * 255).astype(np.uint8)  # 转换为 [0, 255] 范围


# rgb图像
rgb_path = path.replace("depth", "rgb").replace("npz", "png")
rgb = cv2.imread(rgb_path)[:, :, ::-1]

# 显示rgb图像和depth图像,并列放置
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(rgb)
plt.axis("off")
plt.title("RGB")

plt.subplot(1, 2, 2)
plt.imshow(depth_colored, cmap="gray")
plt.axis("off")
plt.title("Depth")

plt.imsave("saved/rgb.png", rgb)
plt.imsave("saved/depth_colored.png", depth_colored)
# plt.imsave("saved/depth.png", depth_normalized)


plt.show()
```
